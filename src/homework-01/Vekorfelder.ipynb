```python
import numpy as np
import matplotlib.pyplot as plt

# Define the electric potential function
def V(x, y):
    return np.sqrt((x+1)**2 + y**2) - np.sqrt((x-1)**2 + y**2)

# Create a meshgrid of coordinates
X, Y = np.meshgrid(np.arange(-4, 4, 0.2), np.arange(-4, 4, 0.2))

# Calculate the electric field components Ex and Ey
Ex = -((X + 1) / ((X+1)**2 + Y**2) - (X - 1) / ((X-1)**2 + Y**2))
Ey = -(Y / ((X+1)**2 + Y**2) - Y / ((X-1)**2 + Y**2))


# Create a new figure and plot the field lines
fig1 = plt.figure()

ax = fig1.add_subplot(1, 1, 1)
ax.spines['bottom'].set_position('center')
ax.spines['left'].set_position('center')
plt.plot(-1, 0, color='green', marker='o', markersize=12, label='q-')
plt.plot(1, 0, color='red', marker='o', markersize=12, label='q+')
plt.streamplot(X, Y, Ex, Ey, density=1.4, linewidth=None, color='grey')
ax.plot([-4, 4], [0, 0], color='grey', linestyle='-', label='Feldlinien')
plt.title('Feldlinien')

plt.xlim([-4, 4])
plt.ylim([-4, 4])

plt.xticks(np.arange(-4, 5), ['']*3 + ['a', '', 'a'] + ['']*3)
plt.yticks(np.arange(-4, 5), ['']*9)




# Add the equipotential lines
levels = np.linspace(-3, 3, 21)  # choose the levels of electric potential
plt.contour(X, Y, V(X, Y), levels=levels, colors='peachpuff', linestyles='dashed', label ="Vekorfeld")
ax.plot([-4, 4], [0, 0], color='peachpuff', linestyle='-', label='Äquipotentiallinien')

plt.legend(loc='upper right', fontsize='small')

plt.savefig('Feldlinien.pdf')
```
